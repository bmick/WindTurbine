```python
import math

def Diff_Power(r):
    # This function returns the differential power generated 
    # by an infinitesimal segment of the blade located at a distance r of the center of the rotor
    
    lift_coef = 1.27
    chord_r = hub_chord * (1.- sigma* ((r**2 - r_hub**2) / (r_tip**2 - r_hub**2)))  # chord at distance r of rotor (m)
    #print chord_r
    wind_speed = math.sqrt ((2.* v1 / 3.)**2 + (omega * r)**2)             # magnitude of wind relative speed (m/s)
    #print wind_speed
    
    dW_dot = 1./ 3. * nb_blade * air_dens * wind_speed * v1 * lift_coef * chord_r * omega * r
    
    #print dW_dot
    return dW_dot
```

```python
import math

def Turbine_power(R, sigma, v1):
    # This function calculates the total power generated by the turbine 
    # with the relation obtained in task 1
    
    r_hat  = r_hub / R
    Lambda = omega * R / v1
    
    Turbine_power = 1. / 9. * nb_blade * air_dens * lift_coef * hub_chord * Lambda**2 * v1**3 * R / (1. - r_hat**2)*\
                    (((4. / (9.*Lambda**2) + 1)**(3./2.)) * (1. - r_hat**2 + sigma*(r_hat**2 - 3./5. + 8./(45.*Lambda**2))) \
                    - (4./(9.*Lambda**2) + r_hat**2)**(3./2.) * (1. - r_hat**2 + sigma*(r_hat**2 - 3./5.*r_hat**2 + 8./(45.*Lambda**2))))
    
    #print "Power output of entire rotor with task 1 formula = ", Turbine_power, "W"
    return Turbine_power
```

```python
def Betz_efficiency(R, sigma, v1):
    # This function calculates the Betz efficiency with the relation obtained in task 1
    
    Betz_eff = 27. * nb_blade * lift_coef * hub_chord * Lambda**2 / (72 * (1. - r_hat**2) * math.pi * R) *\
                    (((4. / (9.*Lambda**2) + 1)**(3./2.)) * (1. - r_hat**2 + sigma*(r_hat**2 - 3./5. + 8./(45.*Lambda**2))) \
                    - (4./(9.*Lambda**2) + r_hat**2)**(3./2.) * (1. - r_hat**2 + sigma*(r_hat**2 - 3./5.*r_hat**2 + 8./(45.*Lambda**2))))
    return Betz_eff
```

```python
def setup_angle (attack_angle, r):
# Returns the setup angle in degree, necessary to keep the alpha angle constant at radius = r
    
    psi = math.atan(2./3. * v1/(omega*r)) #psi in radians
    setup_angle = attack_angle - math.degrees(psi)
    return setup_angle
```
